## Data Collection and Extraction

This step involves collecting data from datasets and then extracting relevant data that would be helpful for our analysis. For the most part the columns of data that we are focusing on are year, state, and event for the storm data. 
For the average temperature data we are focusing on state, year, and average temperature. 


First we have the data of storms and severe weather from 1980 to 2012 in the United States, that details deaths, injuries, where it occurred.

```python
stormData = []

for i in range(1980, 2013):
    if i == 2002:
        input = pd.read_csv("master/StormEvents_details-ftp_v1.0_d{}_c20211102.csv".format(i))
    else:
        input = pd.read_csv("master/StormEvents_details-ftp_v1.0_d{}_c20210803.csv".format(i))
    stormData.append(input)
for i in range(len(stormData)):
    
    print("{} stormData:".format(1980+i))
    display(stormData[i])
```

Now we will extrapolate the data that we want from the Storm Data. Relevant data would be year, state, what kind of weather happened(event). We will do this by dropping the columns that we do not need from the dataset. 

```python
for i in range(len(stormData)):
    stormData[i] = stormData[i].drop(columns=['BEGIN_YEARMONTH','BEGIN_DAY', 
                                              'BEGIN_TIME', 'END_YEARMONTH', 
                                              'END_DAY', 'END_TIME', 
                                              'EPISODE_ID','EVENT_ID', 
                                              'STATE_FIPS','CZ_TYPE', 
                                              'CZ_FIPS','CZ_NAME', 
                                              'WFO','BEGIN_DATE_TIME', 
                                              'CZ_TIMEZONE','END_DATE_TIME', 
                                              'END_RANGE', 'END_AZIMUTH', 
                                              'END_LOCATION', 'BEGIN_LAT', 
                                              'BEGIN_LON', 'END_LAT', 
                                              'END_LON', 'EPISODE_NARRATIVE', 
                                              'EVENT_NARRATIVE', 'DATA_SOURCE', 
                                              'TOR_F_SCALE', 'TOR_LENGTH', 
                                              'TOR_WIDTH', 'TOR_OTHER_WFO', 
                                              'TOR_OTHER_CZ_STATE', 'TOR_OTHER_CZ_FIPS', 
                                              'TOR_OTHER_CZ_NAME', 'BEGIN_RANGE', 
                                              'BEGIN_AZIMUTH', 'BEGIN_LOCATION', 
                                              'SOURCE', 'MAGNITUDE', 
                                              'MAGNITUDE_TYPE', 'CATEGORY',
                                              'INJURIES_DIRECT', 'INJURIES_INDIRECT', 
                                              'DEATHS_DIRECT', 'DEATHS_INDIRECT', 
                                              'DAMAGE_PROPERTY', 'DAMAGE_CROPS', 
                                              'FLOOD_CAUSE'
                                             ])
    print("{} stormData:".format(1980+i))
    display(stormData[i])
```

Now we will define severe weather as weather that is related temperature and extract that data, which is Hurricanes, tornadoes, Excessive Heat, and Extreme Cold/Wind Chill. We then will combine the data extracted into one dataframe.

```python
data = pd.concat(stormData)
tData = data[data['EVENT_TYPE'] == 'Tornado']
hData = data[data['EVENT_TYPE'] == 'Hurricane (Typhoon)']
ehData = data[data['EVENT_TYPE'] == 'Excessive Heat']
ecData = data[data['EVENT_TYPE'] == 'Extreme Cold/Wind Chill']


td = tData.groupby(['YEAR','STATE']).size().to_frame(name = 'numTornado').reset_index()
hd = hData.groupby(['YEAR', 'STATE']).size().to_frame(name = 'numHurricane').reset_index()
ehd = ehData.groupby(['YEAR', 'STATE']).size().to_frame(name = 'numHeat').reset_index()
ecd = ecData.groupby(['YEAR', 'STATE']).size().to_frame(name = 'numCold').reset_index()

merged = pd.merge(left = td, right = hd,how = 'left', on = ['YEAR', 'STATE'])
merged2 = pd.merge(left = merged, right = ehd, how = 'left',  on = ['YEAR', 'STATE'])
mergedFinal = pd.merge(left = merged2, right = ecd, how = 'left', on = ['YEAR', 'STATE'])
mergedFinal
```

We will now extract the data from the global temperature csv file and extract data from the United States from 1980 to 2012

```python
tempByState = pd.read_csv("Globaltempbycountry/GlobalLandTemperaturesByState.csv") 
tempByState.sort_values(by = 'dt')
usTemp = tempByState[tempByState['Country'] == 'United States']
usTemp['dt'] = pd.to_datetime(usTemp['dt'])
usTemp = usTemp.set_index('dt')
temp = usTemp.sort_index().truncate(before = '1980-01-01', after = '2012-12-01')

temp
```

The next step is to find the average temperature per year in the United States

```python
avg = temp.reset_index()
avg['Year'] = avg['dt'].dt.year
avgTemp = avg.groupby(['Year', 'State']).mean()
avgTemp
```

## Data Analysis and Visualization

We are now going to visualize the data by plotting the data into line graphs. For the storm data we split the graphs into 4 different line graphs with the first corresponding to Tornado Frequency, the next being Hurricane Frequency, then Excessive Heat frequency, and last Extreme Cold frequency. 

We also plotted the data for average temperature per year for each state. The purpose of the graphs is to see what kind of correlations there are between storm data and average temperature. This will help us to confirm the various studies talked about in the news relating rising temperatures with more severe weather. 

Something to note, the lines in the graphs are not labelled due to it becoming too cluttered to see any data effectively. However, this does not affect our analysis too much as we can focus on specific states and see what kind of trends they have, such as increasing every year, or whether there are many states that increase in frequency of severe weather every year. 

```python
states = mergedFinal['STATE'].drop_duplicates()

mergedFinal = mergedFinal.set_index('STATE')
for i in states:
    if i in mergedFinal.index:
        plt.plot(mergedFinal.loc[i, 'YEAR'], mergedFinal.loc[i, 'numTornado'])
        
plt.title('Number of Tornadoes per year for each state')
plt.xlabel('Years')
plt.ylabel('Number of Tornadoes')
```

```python
for i in states:
    if i in mergedFinal.index:
        plt.plot(mergedFinal.loc[i, 'YEAR'], mergedFinal.loc[i, 'numHurricane'])
plt.title('Number of Hurricanes per year for each state')
plt.xlabel('Years')
plt.ylabel('Number of Hurricanes')
```

```python
for i in states:
    if i in mergedFinal.index:
        plt.plot(mergedFinal.loc[i, 'YEAR'], mergedFinal.loc[i, 'numHeat'])
plt.title('Number of Excessive Heat per year for each state')
plt.xlabel('Years')
plt.ylabel('Number of Excessive Heat')
```

```python
for i in states:
    if i in mergedFinal.index:
        plt.plot(mergedFinal.loc[i, 'YEAR'], mergedFinal.loc[i, 'numCold'])
plt.title('Number of Extreme Cold per year for each state')
plt.xlabel('Years')
plt.ylabel('Number of Extreme Cold')
```

```python
avgTemp = avgTemp.reset_index()
state = avgTemp['State'].drop_duplicates()
avgTemp = avgTemp.set_index('State')

for i in state:
    if i in avgTemp.index:
        plt.plot(avgTemp.loc[i, 'Year'], avgTemp.loc[i, 'AverageTemperature'])
plt.title('Average Temperature in Celcius per Year in Each State')
plt.xlabel('Years')
plt.ylabel('Average Temperature')
```

```python
avgTempF = avgTemp
avgTempF['AverageTemperature'] = (avgTemp['AverageTemperature'] *1.8) +32
for i in state:
    if i in avgTempF.index:
        plt.plot(avgTempF.loc[i, 'Year'], avgTempF.loc[i, 'AverageTemperature'])
plt.title('Average Temperature in Fahrenheit per Year in Each State')
plt.xlabel('Years')
plt.ylabel('Average Temperature')
```

Unfortunately due to the low amount of data on the hurricane frequency we can not effectively conclude any correlation between hurricanes and rising average temperatures. 

### Extreme heat frequency vs Average Temperature

Similarly to the Hurricane graph there is not much data to work with however in the last couple years in the Extreme heat graph we can see large increases every year in the amount of times there was an extreme amount of heat. This also corresponds with the average temperature graphs as we can see large increases of a couple degrees in the last couple of years. 

### Extreme cold frequencey vs Average Temperature

Unfortunately there is not enough data to conclusively tell whether there is a correlation between extreme cold and average temperature increase. It is possible that due to rising temperatures that the number of times the weather has been extremely cold occurs less often, however this cannot be said for sure with only a couple years worth of extreme cold frequency data
